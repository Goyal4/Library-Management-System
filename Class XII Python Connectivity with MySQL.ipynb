Understand Python MySQL Select query program

1. import mysql.connector  this line imports the MySQL Connector Python module in your program so you can use the methods of this module to communicate with the MySQL database.

2. Next, We used the mysql.connector.connect()  methods to connect the MySQL Database from Python.

3. Then, We prepared SQL SELECT query. SELECT statement returns a set of rows which is called a result set. So we are populating the cursor by executing the SELECT statement.

4. Next, we used a connection.cursor() methods to get a cursor object from the connection object.A SQL cursor is a database object that is used to retrieve data from a result set. 

5. We executed the select operation using a execute() method of a Cursor object.On successful execution of a select query, the execute()  method returns a ResultSet which contains all the rows. 
 It is used to execute the SQL statements with the database cursor object.
 
 After that, We are using a cursor.fetchall()  method we can fetch all the records present under a  table.

6. Once all record fetched successfully, we closed the MySQL Cursor object using a cursor.close() and MySQL database connection using the mySQLconnection.close()


```python
import mysql.connector as mc
con=mc.connect(user="root",password="1234",host="localhost",database='library')
if con.is_connected(): 
    cur=con.cursor()
    cur.execute("select * from manas")
    print(cur.rowcount)
    x=cur.fetchall()
    print(cur.rowcount)
    #x=cur.fetchall()
    print(type(x),x)
    for i in x:
        print(i)
else:
    print("sorry can't connected")
```

After executing a query, a MySQLCursorBuffered cursor fetches the entire result set from the server and buffers the rows. For queries executed using a buffered cursor, row-fetching methods such as fetchone() return rows from the set of buffered rows. 
Using buffered cursor, you can get the number of rows returned by checking MySQLCursorBuffered.rowcount. 

For nonbuffered cursors, rows are not fetched from the server until a row-fetching method is called.
That is why the rowcount attribute of an unbuffered cursor returns -1 right after the execute method is called. 

Furthermore, the rowcount attribute of an unbuffered cursor increases as you fetch rows from it, while the rowcount attribute of a buffered cursor remains the same, as you fetch rows from it.


```python
import mysql.connector
con=mysql.connector.connect(user="root",password="123",host="localhost",database="test")
if con.is_connected(): 
    cur=con.cursor()
    cur.execute("select * from student")
    print(cur.rowcount)
    
    x=cur.fetchone()
    print(x)
    print(cur.rowcount)
    x=cur.fetchone()
    print(x)
    print(cur.rowcount)
```

```python
import mysql.connector
con=mysql.connector.connect(user="root",password="123",host="localhost",database="test")
if con.is_connected(): 
    cur=con.cursor(buffered=True)
    cur.execute("select * from student")
    print(cur.rowcount)
    cur.fetchone()
    print(cur.rowcount)
else:
    print("sorry can't connected")
```

In an unbuffered cursor, you must be sure to fetch all rows of the result set before executing any other statements on the same connection, or an InternalError (Unread result found) exception will be raised.

What it is basically saying is that the result returned by the unbuffered cursor needs to be fully consumed before you can execute another query with any cursor under the same connection.

```python
#example demonstrating Internal Error using unbuffered cursor
import mysql.connector
con=mysql.connector.connect(user="root",password="123",host="localhost",database="test")
if con.is_connected(): 
    cur=con.cursor()
    cur.execute("select * from student")
    y=cur.fetchone()
    print(y)
    print(cur.rowcount)
    cur.execute("select name from student")
    x=cur.fetchall()
    print(x)
else:
    print("sorry can't connected")
```
